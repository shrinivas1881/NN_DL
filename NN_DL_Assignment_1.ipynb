Use dataset available with name glass

```python
# Importing pandas library

import pandas as pd

import warnings
warnings.filterwarnings("ignore")

# Reading the glass dataset

df_glass=pd.read_csv('glass.csv')

print(df_glass.head())

# Checking for the number of rows and columns
df_glass.shape
```

```python
# Checking if any null values in the data
df_glass.isnull().sum()
```

```python
# Checkin unique value counts in the data
df_glass.nunique(axis=0)
```

```python
# Importing matplotlib and seaborn library
import matplotlib.pyplot as plt
import seaborn as sns

print(df_glass.corr())

# plotting correlation heatmap
dataplot = sns.heatmap(df_glass.corr(), cmap="YlGnBu", annot=True)

# displaying heatmap
plt.show()
```

```python
# X -> features, y -> label
X=df_glass.loc[:, df_glass.columns != 'Type']
y=df_glass.loc[:, df_glass.columns == 'Type']
```

```python
from sklearn.model_selection import train_test_split
# dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

# training a Naive Bayes classifier
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)

# accuracy on X_test
accuracy = gnb.score(X_test, y_test)
print(accuracy)
```

```python
# Importing library for classification report
from sklearn.metrics import classification_report
# Classification report for Naives bayes
print(classification_report(y_test, gnb_predictions))
```

3. Implement Linear Regression using scikit-learn

a) Import the given ‚ÄúSalary_Data.csv‚Äù

b) Split the data in train_test partitions, such that 1/3 of the data is reserved as test subset.

c) Train and predict the model.

d) Calculate the mean_squared error.

e) Visualize both train and test data using scatter plot

```python
# Reading the salary dataset

df_salary=pd.read_csv('Salary_Data.csv')

print(df_salary.head())

# Checking for the number of rows and columns
df_salary.shape
```

```python
# Checking if any null values in the data
df_salary.isnull().sum()
```

```python
# Checkin unique value counts in the data
df_salary.nunique(axis=0)
```

```python
# Show statistical description of the data
print(df_salary.info())
df_salary.describe()
```

```python
# Importing matplotlib and seaborn library
import matplotlib.pyplot as plt
import seaborn as sns

print(df_salary.corr())

# plotting correlation heatmap
dataplot = sns.heatmap(df_salary.corr(), cmap="YlGnBu", annot=True)

# displaying heatmap
plt.show()
```

```python
# Creating a scatter plot for the YearsExperience and Salary columns
df_salary.plot.scatter(x="YearsExperience",y="Salary")
```

```python
# X -> features, y -> label
X=df_salary.loc[:, df_salary.columns != 'Salary']
y=df_salary.loc[:, df_salary.columns == 'Salary']
```

```python
# dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=101)
```

```python
# training a linear regression model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)

predictions = lm.predict(X_test)
print(predictions)

# model accuracy for X_test
accuracy = lm.score(X_test, y_test)
print(accuracy)
```

```python
# calculating Mean Squared Error
# Error using sklearn
from sklearn.metrics import mean_squared_error ,mean_absolute_percentage_error
# calculate MSE
print("Mean Squared error : " + str(mean_squared_error(y_test,predictions)))
# calculate MAPE
print("Mean absolute percentage error : " + str(mean_absolute_percentage_error(y_test,predictions)))
```

```python
# Plotting Y_test and Predictions

# Create a list of indices for x-axis
x_axis = range(len(y_test))

# Plot real values and predictions
plt.plot(x_axis, y_test, label='y_test')
plt.plot(x_axis, predictions, label='predictions')

# Add labels and title
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('y_test vs Predictions')

# Add legend
plt.legend()

# Display the plot
plt.show()

```
